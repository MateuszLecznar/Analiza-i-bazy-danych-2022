```python
import pandas as pd
import matplotlib.pyplot as plt

#Czyszczenie danych

pd.set_option('display.max_rows', None)


df = pd.read_csv('14_WIELKOPOLSKIE.csv')
df.drop( columns={'Unnamed: 0':'Index'}, inplace=True )
df.dropna()
df = df[(df['Płeć kupującego'] =="M")|(df["Płeć kupującego"]=="K")]
df=df[df['Wiek kupującego']>0]
df=df[df['Ocena']>=0]  

df.to_csv("CLear_data")

df.head()
```

Obróbka danych

```python


df1=df["Marka"].value_counts()
df1.plot(kind="bar",legend=None)
plt.title("Ilość odkurzaczy kupionych zależnie od marki")
plt.show()
```

```python
df1=pd.DataFrame(df['Płeć kupującego'].value_counts())
df1.plot(kind="bar",legend=None)
plt.title("Ilość odkurzaczy kupionych przez różne płcie")
plt.show()
```

```python

df2=pd.DataFrame(df['Wiek kupującego'])
df2.plot(kind="kde",legend=None)
plt.title('wiek kupujących')
plt.show()
```
